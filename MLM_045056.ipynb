```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.preprocessing import StandardScaler

# Load the dataset
data = pd.read_csv("/content/cleaned_global_superstore_no_missing.csv")

# Assuming 'sales' and 'profit' as features for clustering
X = data[['sales', 'profit']]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Elbow Curve
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.show()

# K-means clustering with k=2
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_scaled)
data['cluster'] = kmeans.labels_

# Plot the data points and cluster centers
plt.scatter(X['sales'], X['profit'], c=data['cluster'], cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('K-means Clustering with k=2')
plt.show()

# Davies-Bouldin Score
db_score = davies_bouldin_score(X_scaled, kmeans.labels_)
print("Davies-Bouldin Score:", db_score)

# Silhouette Score
silhouette_avg = silhouette_score(X_scaled, kmeans.labels_)
print("Silhouette Score:", silhouette_avg)
```

The DB score quantifies the compactness and separation of clusters: lower values indicate better clustering performance, with values closer to 0 indicating well-separated and compact clusters. After computing DB scores for k=2, k=3, k=4, and k=5, we observe that the DB score is highest for k=3, with a value of 0.7306. This suggests that the clustering with three clusters achieves the best balance between intra-cluster similarity and inter-cluster dissimilarity. Hence, based on the DB scores, we choose to partition the data into three clusters for subsequent analysis.

```python
from scipy.stats import f_oneway

# Assuming 'data' is your DataFrame
# Assuming 'kmeans' is your KMeans model with k=3

# Joining cluster labels with the original dataset
data_with_clusters = data.copy()
data_with_clusters['Cluster_Label'] = kmeans.labels_

# Extracting non-categorical variables relevant to your dataset
non_cat_variables = ['sales', 'profit', 'discount', 'shipping_cost', 'quantity']

# Grouping variables by cluster label
cluster_groups = data_with_clusters.groupby('Cluster_Label')

# Perform ANOVA for each non-categorical variable
anova_results_non_cat = {}
for column in non_cat_variables:
    # Collecting groups for the ANOVA test
    groups = [group[column].dropna() for name, group in cluster_groups]
    # Ensure there are at least two groups with non-null values to compare
    if len(groups) > 1 and all(group.size > 0 for group in groups):
        anova_result = f_oneway(*groups)
        anova_results_non_cat[column] = anova_result

# Print ANOVA results for non-categorical variables
for column, result in anova_results_non_cat.items():
    print(f"Variable: {column}")
    print(f"F-value: {result.statistic}")
    print(f"P-value: {result.pvalue}")
    print()
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

# Assuming 'data' is your DataFrame, and it now includes a 'Cluster_Label' column from KMeans clustering
# Let's first add the cluster labels to your DataFrame if you haven't done so
# Assuming 'kmeans' is your KMeans model object after fitting

data['Cluster_Label'] = kmeans.labels_

# Extracting categorical variables from your dataset
cat_variables = ['category', 'country', 'segment']  # Assuming these are categorical variables from your dataset

# Initialize lists to store Chi-square test results for categorical variables
chi2_values = []
p_values = []

# Perform Chi-square test for each categorical variable
for column in cat_variables:
    if column in data.columns:
        # Creating a contingency table between each categorical variable and the cluster labels
        contingency_table = pd.crosstab(data[column], data['Cluster_Label'])
        # Performing the Chi-square test
        chi2, p_value, _, _ = chi2_contingency(contingency_table)
        # Storing the results
        chi2_values.append(chi2)
        p_values.append(p_value)

# Print Chi-square test results for categorical variables
for i, column in enumerate(cat_variables):
    print(f"Variable: {column}")
    print(f"Chi-square: {chi2_values[i]}")
    print(f"P-value: {p_values[i]}")
    print()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Assuming 'data' is your DataFrame
# Selecting and standardizing the data for clustering
features = data[['sales', 'profit']]
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Perform K-means clustering for k=3
kmeans = KMeans(n_clusters=3, random_state=111)
kmeans.fit(features_scaled)

# Add the cluster labels to your DataFrame
data['Cluster_Label'] = kmeans.labels_

# Plot the scatter plot with clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='sales', y='profit', hue='Cluster_Label', legend='full', palette='viridis')

# Plot centroids
# Note: Centroid coordinates need to be inversely transformed if original data was scaled
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
plt.scatter(centroids[:, 0], centroids[:, 1], s=100, c='red', label='Centroids')

# Set plot title and labels
plt.title('Clusters with Centroids for k=3')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.legend()
plt.grid(True)

plt.show()
```
